## [Points: 10] Exercise 2 (NERC): feature inspection using the [Annotated Corpus for Named Entity Recognition](https://www.kaggle.com/abhinavwalia95/entity-annotated-corpus)
**[6 points] a. Perform the same steps as in the previous exercise. Make sure you end up for both the training part (*df_train*) and the test part (*df_test*) with:**
* the features representation using **DictVectorizer**
* the NERC labels in a list

Please note that this is the same setup as in the previous exercise:
* load both train and test using:
    * list of dictionaries for features
    * list of NERC labels
* combine train and test features in a list and represent them using one hot encoding
* train using the training features and NERC labels

````python
# Imports
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
````

````python
# Load dataset
path = "ner_dataset.csv"

kaggle_dataset = pd.read_csv(
    path,
    encoding="latin1",
    on_bad_lines="skip"
)
````

Inspect dataset

````python
len(kaggle_dataset)
````

````python
kaggle_dataset.head()
````

````python
kaggle_dataset.columns
````

````python
# Split dataset
df_train = kaggle_dataset[:100000]
df_test = kaggle_dataset[100000:120000]
print(len(df_train), len(df_test))
````

````python
# Training features and labels
training_features = []
training_gold_labels = []

for _, row in df_train.iterrows():
    feature_dict = {
        "word": row["Word"],
        "pos": row["POS"]
    }

    training_features.append(feature_dict)
    training_gold_labels.append(row["Tag"])
````

````python
# Test features and labels
test_features = []
test_gold_labels = []

for _, row in df_test.iterrows():
    feature_dict = {
        "word": row["Word"],
        "pos": row["POS"]
    }

    test_features.append(feature_dict)
    test_gold_labels.append(row["Tag"])
````

````python
print(training_features[:5])
print(training_gold_labels[:5])
````

````python
# Vectorization using one-hot encoded features
vec = DictVectorizer()

all_features = training_features + test_features
feature_matrix = vec.fit_transform(all_features)

# Split back into training and test matrices
train_array = feature_matrix[:len(training_features)]
test_array = feature_matrix[len(training_features):]
````

````python
print(train_array.shape)
print(test_array.shape)
````

**[4 points] b. Train and evaluate the model and provide the classification report:**
* use the SVM to predict NERC labels on the test data
* evaluate the performance of the SVM on the test data

Analyze the performance per NERC label.

````python
# Imports
from sklearn import svm
from sklearn.metrics import classification_report
````

````python
# Train SVM
linear_svm = svm.LinearSVC(max_iter=10000)

linear_svm.fit(train_array, training_gold_labels)
````

````python
predictions = linear_svm.predict(test_array)

# Classification Report
print(classification_report(test_gold_labels, predictions, zero_division=0))
````

Performance analysis

The classifier performs best on the O label, with an F1-score of 0.99. This is expected, since most tokens in the dataset belong to this class. The model also performs well on labels such as B-gpe (0.94 F1), B-tim (0.83 F1), and B-geo (0.78 F1). I-nat also has a high score, although this is less meaningful statistically because there are only 4 examples in the test set.

Performance is noticeably lower for labels such as B-org (0.57 F1), I-org (0.53 F1), and especially I-tim (0.14 F1). For I-tim, the recall is very low, meaning the model often misses words that are part of longer time expressions. I-per shows a different issue: recall is high (0.90), but precision is much lower (0.42), meaning that the classifier predicts this label too often and makes many false positives.

The rarest labels, such as B-art and B-eve, receive scores of 0.00 because there are very few or no examples for the classifier to learn from. Overall, the results show that the classifier performs much better on common labels than on rare ones. Even though the overall accuracy is high at 0.94, the macro average F1-score of 0.52 shows that performance differs quite a lot across the individual NERC labels.
