````python
def vader_output_to_label(vader_output):
    """
    map vader output e.g.,
    {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': 0.4215}
    to one of the following values:
    a) positive float -> 'positive'
    b) 0.0 -> 'neutral'
    c) negative float -> 'negative'
    
    :param dict vader_output: output dict from vader
    
    :rtype: str
    :return: 'negative' | 'neutral' | 'positive'
    """
    compound = vader_output['compound']
    
    if compound < 0:
        return 'negative'
    elif compound == 0.0:
        return 'neutral'
    elif compound > 0.0:
        return 'positive'
    
assert vader_output_to_label( {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': 0.0}) == 'neutral'
assert vader_output_to_label( {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': 0.01}) == 'positive'
assert vader_output_to_label( {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': -0.01}) == 'negative'
````

````python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report

analyzer = SentimentIntensityAnalyzer()

tweets = []
all_vader_output = []
gold = []

# settings (to change for different experiments)
to_lemmatize = True 
pos = set()

for id_, tweet_info in my_tweets.items():
    the_tweet = tweet_info['text_of_tweet']

    # run vader
    vader_output = analyzer.polarity_scores(the_tweet)
    # convert vader output to category
    vader_label = vader_output_to_label(vader_output)

    tweets.append(the_tweet)
    all_vader_output.append(vader_label)
    gold.append(tweet_info['sentiment_label'])
    
# use scikit-learn's classification report
print(classification_report(gold, all_vader_output))
````

a. Quantitative evaluation
The scores are uneven across the three sentiment categories. Negative tweets have high precision (0.93) which means that when classifier makes a prediction, it is usually a correct one. However, the recall is lower (0.68), which shows that some negative tweets are not detected.
The performance on neutral tweets is very poor, with a very low recall (0.19) and moderate precision (0.50). This indicates that the classifier struggles to correctly identify neutral tweets and often misclassifies them as either positive or negative.
For positive tweets, the recall is very high (1.00), meaning that the classifier identifies all positive tweets. However, the precision is only 0.50, which means that many tweets predicted as positive are actually not positive.

The macro average F1-score (0.58) shows that performance is not balanced across the different categories. The overall accuracy is 0.62, which is moderate, but does not reveal how poorly it performed on netural tweets. Overall, the classifier tends to over-predict positive sentiment and has difficulty correctly identifying neutral tweets.

b. Error analysis

There are 19 misclassified tweets in total (13 neutral and 6 negative). There were no misclassified positive tweets which is consistent with the recall for the positive category.

Most neutral tweets are predicted as positive. Many of these tweets are factual or informational, such as “This is why OpenAI was created”, the NCAA announcement, or “I’m joining X to share occasional updates...”. VADER assigns positive sentiment to certain words like “created”, “welcome”, or “on track” based on its lexicon, even though the overall meaning of the sentence is neutral. There is also one case where a neutral tweet is predicted as negative, likely due to negative percentages and capitalized phrases like “NO ONE IS TALKING…”. In the VADER lexicon, these elements may receive negative scores, even though the tweet itself is more informational.

When it comes to negative tweets, some are predicted as neutral, such as “They stole a nonprofit. It’s not right.”. This tweet does not contain words with strong negative scores in the VADER lexicon, even though the overall meaning is clearly negative. This shows that VADER struggles when sentiment depends more on context than on explicit sentiment words.

Many negative tweets are classified as positive, such as “Even though the sea is so enchanting and wonderful, seafaring is a terrifying profession.”. In this case, VADER gives too much weight to positive words like “enchanting” and “wonderful” and fails to capture the overall negative sentiment.

Some cases are also ambiguous, such as news reports about negative events like flooding. While these can be interpreted as negative due to the situation described, they can also be seen as neutral because they are presented in a factual way. This makes classification more difficult for both humans and the model.

Generally, many misclassified tweets are news-related or factual. VADER assigns sentiment based on individual words from its lexicon and simple rules, but it often fails to capture the overall contextual meaning of the sentence.

````python
def important_features_per_class(vectorizer,classifier,n=80):
    class_labels = classifier.classes_
    feature_names =vectorizer.get_feature_names()
    topn_class1 = sorted(zip(classifier.feature_count_[0], feature_names),reverse=True)[:n]
    topn_class2 = sorted(zip(classifier.feature_count_[1], feature_names),reverse=True)[:n]
    topn_class3 = sorted(zip(classifier.feature_count_[2], feature_names),reverse=True)[:n]
    print("Important words in negative documents")
    for coef, feat in topn_class1:
        print(class_labels[0], coef, feat)
    print("-----------------------------------------")
    print("Important words in neutral documents")
    for coef, feat in topn_class2:
        print(class_labels[1], coef, feat) 
    print("-----------------------------------------")
    print("Important words in positive documents")
    for coef, feat in topn_class3:
        print(class_labels[2], coef, feat) 

# example of how to call from notebook:
#important_features_per_class(airline_vec, clf)
````
